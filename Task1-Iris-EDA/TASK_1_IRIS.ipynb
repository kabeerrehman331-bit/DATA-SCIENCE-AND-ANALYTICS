# Task 1: Exploring and Visualizing the Iris Dataset

## Introduction
The Iris dataset contains 150 samples of iris flowers with 4 features:
sepal length, sepal width, petal length, and petal width.
The goal is to explore and visualize the data to understand patterns across 3 species.


```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
print("Shape:", df.shape)
```

```python
print("Columns:", df.columns.tolist())
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', palette='Set1')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

print(df)
```

```python
df.hist(figsize=(10, 6), color='steelblue', edgecolor='black')
plt.suptitle('Distribution of Iris Features', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='species', y='petal_length', palette='Set2')
plt.title('Petal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

## Conclusion

- The dataset has 150 rows and 5 columns with no missing values.
- Setosa is clearly separated from the other two species based on petal length.
- Versicolor and Virginica overlap slightly, making them harder to distinguish.
- Petal length is the most useful feature for separating species.
- No outliers were detected in the box plots.
